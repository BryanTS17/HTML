```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
novos_usuarios = [100, 80, 110, 130, 130, 160, 170, 120, 110, 150, 200, 140]

fig = plt.figure(figsize=(13,5))

plt.plot(meses, novos_usuarios)  # Cria uma linha conectando os pontos (1, 4), (2, 5) e (3, 6)

plt.savefig("" , dpi=150)
plt.show()
```

```python
receita_mensal = [10000, 7000, 9800, 20000, 17000]
meses = ["2021", "2022", "2023", "2024", "2025"]

fig = plt.figure(figsize=(13,5.1))
plt.bar(meses, receita_mensal)  # Plota barras com alturas baseadas nos valores
plt.title("Receita Mensal", fontsize=16)
plt.ylabel("média da receita", fontsize=12)

plt.savefig(r"C:\Users\CPU\Desktop\codigos\html do matplotlib_pyplot\imagens\novo_grafico_barras.png", dpi=65)
plt.show() 
```

```python

quantidades_vendidas = [45, 30, 15, 10]
nomes = ["batata", "batata frita", "pizza", "queijo"]

fig = plt.figure(figsize=(13,5.1))

plt.pie(quantidades_vendidas, labels=nomes)
plt.title("Divisão do orçamento por setores", fontsize=14)

plt.savefig(r"C:\Users\CPU\Desktop\codigos\html do matplotlib_pyplot\imagens\novo_grafico_pie.png", dpi=65)
plt.show()
```

```python

fig = plt.figure(figsize=(13,5.1))
plt.hist([1, 2, 2, 3, 3, 3, 4, 5, 5], bins=5)  # Agrupa os valores em 5 intervalos
plt.title("Histograma", fontsize=16)
plt.savefig(r"C:\Users\CPU\Desktop\codigos\html do matplotlib_pyplot\imagens\novo_grafico_hist.png", dpi=65)
plt.show()
```
